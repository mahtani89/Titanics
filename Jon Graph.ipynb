```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_training=pd.read_csv('/Users/deepakmahtani/.kaggle/competitions/titanic/train.csv')
```

```python
raw_training.columns
```

```python
clean_training=raw_training[pd.notnull(raw_training['Cabin'])][['Survived','Cabin']]
```

```python
clean_training['Cabin'].iloc[0][0]
```

```python
clean_training
```

```python
letter=[]
for index, row in clean_training.iterrows():
    letter.append(row['Cabin'][0])
    
```

```python
clean_training['letter']=letter
```

```python
print(clean_training.groupby(by='letter').mean())
```

```python
clean_training.groupby(by='letter').mean().plot(kind='bar')
```
